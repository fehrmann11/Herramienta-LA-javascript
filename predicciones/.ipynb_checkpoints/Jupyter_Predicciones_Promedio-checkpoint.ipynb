```python
import urllib.request as ur,json,pandas as pd

#conexión al backend, si se pone http://localhost:3000/predProm/número da el curso de moodle al que entramos.
url = "http://localhost:3000/predProm"
response = ur.urlopen(url)
data = json.loads(response.read())

promedio_Y = data[0]
cantidad_foros = data[1]
cantidad_cursos_alumno = data[2]
cantidad_visitas_curso = data[3]
cantidad_notas = data[4]
cantidad_Acciones = data[5]
```

```python
#es normalize ya que viene de una varibale
df_promedio_Y = pd.json_normalize(promedio_Y)
df_cantidad_foros = pd.json_normalize(cantidad_foros)
df_cantidad_cursos_alumno = pd.json_normalize(cantidad_cursos_alumno)
df_cantidad_visitas_curso = pd.json_normalize(cantidad_visitas_curso)
df_cantidad_notas = pd.json_normalize(cantidad_notas)
df_cantidad_Acciones = pd.json_normalize(cantidad_Acciones)
```

```python
df_cantidad_cursos_alumno= df_cantidad_cursos_alumno.drop("rownum",1)
```

```python
df_cantidad_cursos_alumno
```

```python
df_cantidad_cursos_alumno["userid"].value_counts()
```

```python
#de la función value_counts lo transformo en dataframe. 
df_cantidad_cursos_alumno = df_cantidad_cursos_alumno["userid"].value_counts().reset_index()
df_cantidad_cursos_alumno.columns = ['userid','cantidad_cursos']
df_cantidad_cursos_alumno
```

```python
#cantidad de notas
df_cantidad_notas['c_notas'][0]
```

```python
#Proceso de unión de dataframe, a la vez de resetear indices. 

df_brigido = df_cantidad_visitas_curso.set_index('userid').join(df_cantidad_foros.set_index('userid'))
df_brigido = df_brigido.reset_index()
df_brigido = df_brigido.set_index('userid').join(df_cantidad_cursos_alumno.set_index('userid'))
df_brigido = df_brigido.reset_index()
df_brigido = df_brigido.set_index('userid').join(df_cantidad_Acciones.set_index('userid'))
df_brigido = df_brigido.reset_index()
df_brigido = df_brigido.set_index('userid').join(df_promedio_Y.set_index('userid'))
df_brigido = df_brigido.reset_index()
df_brigido
```

```python
#unión de tabla número de notas
df_brigido['numero_notas'] = df_cantidad_notas['c_notas'][0]
df_brigido
```

```python
#rellenar valores null
df_brigido['prom'] =df_brigido['prom'].fillna(0)
df_brigido['contador'] =df_brigido['contador'].fillna(0)
df_brigido
```

```python
import numpy as np
#elimino dataframe y además reinicio el indice.

df_brigido = df_brigido.drop(df_brigido[df_brigido['prom']==0.0].index)
df_brigido.index = np.arange(0, len(df_brigido))
df_brigido
```

```python
#dividimos en sets de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state = 25)
```

```python
#función para la variable y, retornando a partir de la multiplicación de coeficientes
def predict_de(cantidad_visitas,cantidad_foros,cantidad_cursos,acciones,n_notas):
    return  -0.0774 * cantidad_visitas + cantidad_foros * 7.8366  + cantidad_cursos* -54.8553 + acciones * 0.1358  + n_notas*2.8919 
```
